Exploratory Data Analysis

```python
# Create new column classified based on length of time product sold
def kesimpulan (row):
   if row['lama_produk_terjual'] < 31 :
      return 'Diminati'
   if row['lama_produk_terjual'] > 30:
      return 'Cukup Diminati'
   if row['lama_produk_terjual'] > 90:
      return 'Kurang Diminati'
   if row['lama_produk_terjual']  == 0:
      return 'Tidak Diminati'
   return 'Tidak Diminati'
```

```python
dept.apply (lambda row: kesimpulan(row), axis=1)
```

```python
# Apply columns in dataset
dept['Hasil'] = dept.apply (lambda row: kesimpulan(row), axis=1)
```

Data Preparation

```python
# Change data type
dept['created_at'] = pd.to_datetime(dept['created_at'])
dept['sold_at'] = pd.to_datetime(dept['sold_at'])
```

```python
# Check Missing Value
dept.isnull().sum()
```

```python
# Remove missing value Produk Name column with value 'No Name'
dept = dept.fillna({
    'product_name': 'no name'
})
```

```python
# Remove missing value Lama Produk Terjual column with value '0'
dept = dept.fillna({
    'lama_produk_terjual': '0'
})
```

```python
# Remove missing value sold_at column with match date in created_at column
dept = dept.fillna(method='ffill', axis='columns')
```

```python
# Check result
dept.head()
```

```python
# Check again missing value in dataset
dept.isnull().sum()
```

```python
# Check Duplicated
dept.duplicated().sum()
```

```python
# Drop columns
dept = dept.drop(['created_at'], axis=1)
dept = dept.drop(['sold_at'], axis=1)
dept = dept.drop(['latitude'], axis=1)
dept = dept.drop(['longitude'], axis=1)
dept = dept.drop(['product_distribution_center_id'], axis=1)
```

Data Preprocessing

```python
# Change datatype before transform data with labelencoder
dept = dept.astype({'lama_produk_terjual':'str'})
```

```python
#Transormation Data using LabelEncoder
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
dept['id']= le.fit_transform(dept['id'].values)
dept['product_id']= le.fit_transform(dept['product_id'].values)
dept['cost']= le.fit_transform(dept['cost'].values)
dept['product_category']= le.fit_transform(dept['product_category'].values)
dept['product_name']= le.fit_transform(dept['product_name'].values)
dept['product_brand']= le.fit_transform(dept['product_brand'].values)
dept['product_retail_price']= le.fit_transform(dept['product_retail_price'].values)
dept['product_department']= le.fit_transform(dept['product_department'].values)
dept['product_sku']= le.fit_transform(dept['product_sku'].values)
dept['state_name']= le.fit_transform(dept['state_name'].values)
dept['lama_produk_terjual']= le.fit_transform(dept['lama_produk_terjual'].values)
dept['Hasil']= le.fit_transform(dept['Hasil'].values)
```

```python
#Import Library
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.tree import DecisionTreeClassifier
```

```python
# Create variabel x dan y
x = dept.drop(columns='Hasil')
y = dept['Hasil']
```

```python
# Split data
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=1)
tree_product = DecisionTreeClassifier(criterion='entropy', max_depth=2, random_state=1)
tree_product.fit(xtrain,ytrain)
```

```python
# Check accuracy score
ypred = tree_product.predict(xtest)
cm = confusion_matrix(ytest, ypred)
print("Confusion Matrix")
print(cm)
akurasi = classification_report(ytest, ypred)
print("Tingkat Akurasi Algoritma C4.5")
print("Akurasi :", akurasi)
akurasi = accuracy_score(ytest, ypred)
print("Tingkat Akurasi: %d persen" %(akurasi*100))
```
